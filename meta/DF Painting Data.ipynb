```python
import pandas as pd
import json, os
```

```python
# reads all file names in directory, returns them in list

def read_file_names():
    json_name = []

    for filename in os.listdir(os.getcwd()):
        if filename[-4:] == 'json':
            json_name.append(filename)

    return json_name

json_name = read_file_names()
```

```python
# extracts relevant info from each json file
# artistContentId not to be confused with contentId

def extract_info(filenames):
    ID = []
    artist = []
    style = []
    genre = []
    url = []
    
    for i in range(len(filenames)):
        with open(filenames[i]) as f:
            data = json.load(f)
            for j in range(len(data)):
                ID.append(data[j]['contentId'])
                artist.append(data[j]['artistName'])
                style.append(data[j]['style'])
                genre.append(data[j]['genre'])
                url.append(data[j]['image'])
    
    return ID, artist, style, genre, url

ID, artist, style, genre, url = extract_info(json_name)
```

```python
## creating the Pandas Dataframe

def create_df(ID, artist, style, genre, url):
    data = {'id': ID, 'artist': artist, 'style': style, 'genre': genre, 'url': url}
    df = pd.DataFrame(data=data)
    
    return df

df = create_df(ID, artist, style, genre, url)
```

```python
## saving dataframe to pickle

def save_df(df):
    df.to_pickle('meta.p')
    
save_df(df)
```
