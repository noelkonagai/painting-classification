```python
import pandas as pd
import urllib.request
import matplotlib.pyplot as plt
from pylab import rcParams
import os.path
rcParams['figure.figsize'] = 14, 7

metadata = pd.read_pickle('meta.p')
```

```python
for i in range(len(metadata)):
    if metadata['id'][i] == 305776:
        print(metadata['url'][i])
```

```python
def genre_distribution():
    metadata['genre'].value_counts().plot(kind = 'bar', rot = 45)
#     plt.show()

    genres = metadata['genre'].value_counts().index.tolist()
    counts_genre = metadata['genre'].value_counts().tolist()
    
    frq_genres = []
    for i in range(len(counts_genre)):
        if counts_genre[i] >= 100:
            frq_genres.append(genres[i])
            
    return frq_genres

frq_genres = genre_distribution()

frq_genres
```

```python
def style_distribution():
    metadata['style'].value_counts().plot(kind = 'bar', rot = 45)
#     plt.show()

    styles = metadata['style'].value_counts().index.tolist()
    counts_style = metadata['style'].value_counts().tolist()
    
    frq_styles = []
    for i in range(len(counts_style)):
        if counts_style[i] >= 100:
            frq_styles.append(styles[i])
            
    return frq_styles

frq_styles = style_distribution()

frq_styles
```

```python
# paintings that have less ferquent genres but satisfy the style condition, saved in Style

count = 0

for i in range(len(metadata['url'])):
    filename = str(metadata['id'][i]) + ".jpg"
    if os.path.isfile(filename):
        print('file already exists', filename)
        count += 1
    if metadata['style'][i] in frq_styles and metadata['genre'][i] not in frq_genres:
        try:
            urllib.request.urlretrieve(metadata['url'][i], filename)
            count += 1
            print(filename)
        except:
            pass

print(count, len(metadata))
```
